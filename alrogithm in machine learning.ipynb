```python
import pandas as pd 
import numpy as np 
```

```python
recipes = pd.read_csv(r"C:\Users\HP\Desktop\Dataset for analysing\muffins_cupcakes.csv")
```

```python
recipes
```

```python
input_columns= recipes.columns[1:].tolist()
```

```python
input_columns
```

```python
input_values=recipes[input_columns].values.tolist()
```

```python
input_values
```

```python
output = recipes['Type'].values
```

```python
output
```

```python
from sklearn.svm import SVC
```

```python
model = SVC(kernel = 'rbf')
```

```python
model.fit(input_values,output)
```

```python
z=[[55, 28, 3, 7, 5, 2, 0, 0],[34, 17, 23, 11, 13, 0, 1, 0]]
prediction = model.predict(z)

print(prediction)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
model.fit(input_values,output)

h = [[50, 25, 12, 6, 5, 2, 1, 0],[34, 17, 20, 20, 5, 2, 1, 0]]
prediction = model.predict(h)
print(prediction)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
model.fit(input_values,output)

y = [[47, 24, 12, 6, 9, 1, 0, 0],[54, 27, 7, 5, 5, 2, 0, 0],[38, 15, 31, 8, 6, 1, 1, 0]]

prediction = model.predict(y)

print(prediction)
```
